```python
import plotly.express as px
data = {
    'CP + GE: Soft Voting': [0.4360, 0.3446, 0.4252, 0.3948, 0.3304],
    'CP + GE: End to End': [0.4360, 0.3446, 0.4252, 0.3948, 0.3304],
    'CP + CS: Soft Voting': [0.3691, 0.3901, 0.3902, 0.4431, 0.4438],
    'CP + CS: End to End': [0.3691, 0.3901, 0.3902, 0.4431, 0.4438],
    'CS + GE: Soft Voting': [0.4170, 0.3543, 0.4638, 0.4073, 0.3676],
    'CS + GE: End to End': [0.4170, 0.3543, 0.4638, 0.4073, 0.3676],
    'CP + GE + CS: Soft Voting': [0.3987, 0.4223, 0.4659, 0.4138, 0.3403],
    'CP + GE + CS: End to End': [0.3987, 0.4223, 0.4659, 0.4138, 0.3403],
}

import plotly.graph_objects as go

# Convert the data into a pandas DataFrame with long format
long_data = pd.DataFrame(data).melt(var_name='Algorithm', value_name='Performance')

# Add a new 'Group' column to the dataframe
long_data['Group'] = long_data['Algorithm'].str.split(':').str[0]

# Get unique groups and sort them
groups = sorted(long_data['Group'].unique())

# Create an empty figure
fig = go.Figure()

# Add a box plot for each algorithm
for algorithm in long_data['Algorithm'].unique():
    fig.add_trace(go.Box(
        x=long_data['Group'][long_data['Algorithm'] == algorithm],
     
        y=long_data['Performance'][long_data['Algorithm'] == algorithm],
        name=algorithm,
        boxpoints=False,  # do not display the original data points
        whiskerwidth=0.4,
        line=dict(width=2)  # assign each box trace to an algorithm
    ))

# Customize the layout
fig.update_layout(
    title='Performance of 12 Algorithms on erik_hq_8_12',
    yaxis=dict(title='Macro F1 Score'),
    xaxis=dict(title='Algorithm Group', tickangle=45, tickvals=groups, ticktext=groups),
    autosize=False,
    width=1000,
    height=600,
    boxmode='group',  # group together boxes of the different traces for each value of x
    plot_bgcolor='white',  # set the background color to white
    shapes=[
        # add horizontal line
        go.layout.Shape(
            type="line",
            xref="paper", x0=0, x1=1,
            yref="y", y0=0.1, y1=0.1,
            line=dict(color="red", dash="dashdot"),
        ),
        # add vertical lines
        go.layout.Shape(
            type="line",
            xref="x", x0=groups[2], x1=groups[2],
            yref="paper", y0=0, y1=1,
            line=dict(color="blue", dash="solid"),
        ),
        go.layout.Shape(
            type="line",
            xref="x", x0=groups[5], x1=groups[5],
            yref="paper", y0=0, y1=1,
            line=dict(color="green", dash="solid"),
        ),
    ]
)

# Show the plot
fig.show()
```
